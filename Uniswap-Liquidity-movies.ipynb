```python
## The graph pool Liquidity data
import json
from collections import OrderedDict
import pickle
from datetime import datetime
import os

# Enter the appropriate <API_KEY> below

def createMovie(path, basename, address, last=False):

    # function to use requests.post to make an API call to the subgraph url
    def run_query(q):

        # endpoint where you are making the request  <API_KEY>
        request = requests.post('https://gateway.thegraph.com/api/<API_KEY>/subgraphs/id/0x9bde7bf4d5b13ef94373ced7c8ee0be59735a298-2'
                                '',
                                json={'query': q})
        if request.status_code == 200:
            return request.json()
        else:
            raise Exception('Query failed. return code is {}.      {}'.format(request.status_code, q))    
    
    def getPoolData(address, last):
        if last:
            order = 'orderDirection: desc'
            maxI = 1
        else:
            order = 'orderDirection: asc'
            maxI = 100
        swaps = []
        for i in range(maxI):
            qq='{pools(where: {id: "'+address+'"}) {id    swaps(skip: '+str(i*1000)+', first:1000, orderBy: timestamp '+order+' ) {timestamp amount0 amount1 tick}}}'
            r = run_query(qq)
            swaps += r['data']['pools'][0]['swaps']
            print(i, len(r['data']['pools'][0]['swaps']))
            if len(r['data']['pools'][0]['swaps']) == 0:
                break

        mints = []
        for i in range(15):
            qq='{pools(where: {id: "'+address+'"}) {id    mints(skip: '+str(i*1000)+', first:1000, orderBy: timestamp '+order+' ) {timestamp amount tickUpper tickLower}}}'
            r = run_query(qq)
            mints += r['data']['pools'][0]['mints']
            print(i, len(r['data']['pools'][0]['mints']))   
            if len(r['data']['pools'][0]['mints']) == 0:
                break
        burns = []
        for i in range(15):
            qq='{pools(where: {id: "'+address+'"}) {id    burns(skip: '+str(i*1000)+', first:1000, orderBy: timestamp '+order+' ) {timestamp amount  tickUpper tickLower}}}'
            r = run_query(qq)
            burns += r['data']['pools'][0]['burns']
            print(i, len(r['data']['pools'][0]['burns'])) 
            if len(r['data']['pools'][0]['burns']):
                break
        return swaps, mints, burns


    def createLiqFeesArrays(swaps, mints, burns):
        MB = {}
        maxBurnTS = 0
        for i, b in enumerate(burns):
            t = b['timestamp']
            b['amount'] = '-'+b['amount']
            if int(t) > maxBurnTS:
                maxBurnTS = int(t)
            if t not in MB:
                MB[t] = b
            elif str(int(t)+1) not in MB:
                MB[str(int(t)+1)] = b
            elif str(int(t)+2) not in MB:
                MB[str(int(t)+2)] = b
            elif str(int(t)+3) not in MB:
                MB[str(int(t)+3)] = b
            elif str(int(t)+4) not in MB:
                MB[str(int(t)+5)] = b
            elif str(int(t)+5) not in MB:
                MB[str(int(t)+5)] = b  
            elif str(int(t)+6) not in MB:
                MB[str(int(t)+6)] = b
            elif str(int(t)+7) not in MB:
                MB[str(int(t)+7)] = b
            elif str(int(t)+8) not in MB:
                MB[str(int(t)+8)] = b
            elif str(int(t)+9) not in MB:
                MB[str(int(t)+9)] = b
            elif str(int(t)+10) not in MB:
                MB[str(int(t)+10)] = b                  

        maxMintTS = 0
        for i, b in enumerate(mints):
            t = b['timestamp']
            if int(t) > maxMintTS:
                maxMintTS = int(t)    
            if t not in MB:
                MB[t] = b
            elif str(int(t)+1) not in MB:
                MB[str(int(t)+1)] = b
            elif str(int(t)+2) not in MB:
                MB[str(int(t)+2)] = b
            elif str(int(t)+3) not in MB:
                MB[str(int(t)+3)] = b
            elif str(int(t)+4) not in MB:
                MB[str(int(t)+4)] = b     
            elif str(int(t)+5) not in MB:
                MB[str(int(t)+5)] = b
            elif str(int(t)+6) not in MB:
                MB[str(int(t)+6)] = b
            elif str(int(t)+7) not in MB:
                MB[str(int(t)+7)] = b
            elif str(int(t)+8) not in MB:
                MB[str(int(t)+8)] = b                    
            elif str(int(t)+9) not in MB:
                MB[str(int(t)+9)] = b
            elif str(int(t)+10) not in MB:
                MB[str(int(t)+10)] = b   
        SW = {}
        maxSwapTS = 0
        for i, b in enumerate(swaps):
            t = b['timestamp']
            if int(t) > maxSwapTS:
                maxSwapTS = int(t)    
            if t not in SW:
                SW[t] = b
            elif str(int(t)+1) not in SW:
                SW[str(int(t)+1)] = b
            elif str(int(t)+2) not in SW:
                SW[str(int(t)+2)] = b
            elif str(int(t)+3) not in SW:
                SW[str(int(t)+3)] = b
            elif str(int(t)+4) not in SW:
                SW[str(int(t)+4)] = b  
            elif str(int(t)+5) not in SW:
                SW[str(int(t)+5)] = b
            elif str(int(t)+6) not in SW:
                SW[str(int(t)+6)] = b
            elif str(int(t)+7) not in SW:
                SW[str(int(t)+7)] = b
            elif str(int(t)+8) not in SW:
                SW[str(int(t)+8)] = b                  
            elif str(int(t)+9) not in SW:
                SW[str(int(t)+9)] = b
            elif str(int(t)+10) not in SW:
                SW[str(int(t)+10)] = b
            elif str(int(t)+11) not in SW:
                SW[str(int(t)+11)] = b
            elif str(int(t)+12) not in SW:
                SW[str(int(t)+12)] = b  
            elif str(int(t)+13) not in SW:
                SW[str(int(t)+13)] = b
            elif str(int(t)+14) not in SW:
                SW[str(int(t)+14)] = b
            elif str(int(t)+15) not in SW:
                SW[str(int(t)+15)] = b
            elif str(int(t)+16) not in SW:
                SW[str(int(t)+16)] = b                           
            else:
                print(b)

        MB = OrderedDict(sorted(MB.items(), key=lambda t: int(t[0].split('.')[0])))
        SW = OrderedDict(sorted(SW.items(), key=lambda t: int(t[0].split('.')[0])))

   
        k = 0
        maxTick = 887200
        minTick = -887200
        ts = []
        for tim, mb in MB.items():
            tH = int(mb['tickUpper'])
            tL = int(mb['tickLower'])
            ts.append(tim)
            if tH > maxTick:
                maxTick = tH
            if tL < minTick:
                minTick = tL
        
        priceTicks = []
        timeTicks = []
        for tts, r in SW.items():
            priceTicks.append(1.0001**float(r['tick']))
            timeTicks.append(int(tts))
        priceTicks.append(priceTicks[-1])
        timeTicks.append(timeTicks[-1])     
        
        tsA = sort(array(list(set(ts))).astype('int'))
        #tsA = tsA[tsA <= min(maxBurnTS+10, maxMintTS+10)]
        liquidityAll = ones((len(tsA)+1, int((maxTick - minTick)/200)))  

        for mb in list(MB.values()):
            if int(mb['timestamp']) > min(maxBurnTS, maxMintTS):
                break
            tt = (tsA==int(mb['timestamp'])).nonzero()[0][0]
            tLi = int(8872/2 + int(mb['tickLower'])/200)
            tHi = int(8872/2 + int(mb['tickUpper'])/200)+1
            dL = int(mb['amount'])/(tHi-tLi)
            liquidityAll[tt:, tLi:tHi] = liquidityAll[tt:, tLi:tHi] + dL

        feesAll = zeros((len(timeTicks), int((maxTick - minTick)/200)))  
        timeTicksA = array(timeTicks)
        for sw in list(SW.values()):
            tt = (timeTicksA==int(sw['timestamp'])).nonzero()[0][0]
            tLn = min(int(8872/2 + log(priceTicks[tt])/log(1.0001)/200), int(8872/2 + log(priceTicks[tt+1])/log(1.0001)/200))
            tLx = max(int(8872/2 + log(priceTicks[tt])/log(1.0001)/200), int(8872/2 + log(priceTicks[tt+1])/log(1.0001)/200))+1

            dF = abs(float(sw['amount1'])/(tLx-tLn)) + abs(float(sw['amount0'])/(tLx-tLn)*priceTicks[tt])
            feesAll[tt, tLn:tLx] = dF  

        return liquidityAll, feesAll, priceTicks, timeTicks, tsA
    
    print('Loading datafiles?')
    try:
        swaps = load(path+basename+'-swaps.npy', allow_pickle=True)
        mints = load(path+basename+'-mints.npy', allow_pickle=True)
        burns = load(path+basename+'-burns.npy', allow_pickle=True)
        print('Loaded datafiles!')
    except:
        print('No files found, getting remote data')
        swaps, mints, burns = getPoolData(address, last)
        save(path+basename+'-swaps.npy', swaps)
        save(path+basename+'-mints.npy', mints)
        save(path+basename+'-burns.npy', burns)
                

    liquidityAll, feesAll, priceTicks, timeTicks, tsA = createLiqFeesArrays(swaps, mints, burns)
    return  liquidityAll, feesAll, priceTicks, timeTicks, tsA
    def createImgs(path, basename, liquidityAll, feesAll, priceTicks, timeTicks, tsA):
        fig = plt.figure(figsize=(2.5, 1.75), dpi=300, constrained_layout=False)
        ax0 = fig.add_subplot(1, 1, 1)
        j = 0
        pp = 1.0001**(arange(0, 887200*2, 200)-887200)   
        a, b = histogram(log(liquidityAll.reshape(-1, )[log(liquidityAll.reshape(-1, )) > 15]), 100)
        xmin = median(priceTicks[100:-100])/2.5
        xmax = median(priceTicks[100:-100])*2.5
        ymin = (b[0])
        ymax = (b[-1]+1)
        if priceTicks[-1] > 1:
            base=' '+basename
        else:
            base=' ETH'
        print(median(priceTicks[100:-100]))    
        for i in range(0, len(priceTicks)-1):
            try:
                price = priceTicks[i]
            except:
                pass
            semilogx([price, price], [ymin, ymax], lw=0.5, color='#4682b4', label='current price')
            if timeTicks[i] > tsA[j]:
                j = tsA[tsA <= timeTicks[i]].nonzero()[0][-1]+1
                print(j)
            fill_between(pp, log(liquidityAll[j, :]), 1, step='mid', color='#cc333f', alpha=0.25)
            if priceTicks[i] > priceTicks[i-1]:
                fill_between(pp[(pp>priceTicks[i-1])&(pp<priceTicks[i])], log(liquidityAll[j, :])[(pp>priceTicks[i-1])&(pp<priceTicks[i])], 1, step='mid', fc='#cc333f', alpha=0.25, ec='#ffffff')
            else:
                fill_between(pp[(pp>priceTicks[i])&(pp<priceTicks[i-1])], log(liquidityAll[j, :])[(pp>priceTicks[i])&(pp<priceTicks[i-1])], 1, step='mid', fc='#cc333f', alpha=0.25, ec='#ffffff')  
            text(price*1.1, log(1.5*liquidityAll.max()), 'P = '+str(price)[:7], fontsize=4)    
            step(pp, log(liquidityAll[j, :]), color='#cc333f', label='liquidity', where='mid')    
            
            #if base == 'ETH':
            step(pp, ymin+log(0.1+feesAll[:i, :].sum(0))/log(0.1+feesAll[:-1, :].sum(0)).max()*(ymax-ymin)*0.875, color='#35aaaf', label='fees = '+str(feesAll[:i, :].sum()/10**9)[:8]+base.split('-')[0], where='mid')
            #else:
            #    step(pp, ymin+log(0.1+feesAll[:i, :].sum(0))/log(0.1+feesAll[:-1, :].sum(0)).max()*(ymax-ymin)*0.875, color='#35aaaf', label='fees = '+str(0.01*feesAll[:i, :].sum()/priceTicks[-1])[:5]+'ETH')
            text(xmin/2, ymin+1, 'transaction '+str(i), fontsize=5)
            legend(loc=2)
            xlim(xmin/3, xmax)
            ylim((ymin, ymax))
            xlabel(basename.split('-')[0]+' price')
            ylabel('log[Liquidity]')
            title(datetime.utcfromtimestamp(timeTicks[i]).strftime('%Y-%m-%d %H:%M:%S'))
            plt.draw()
            tight_layout()
            plt.savefig(path+basename+f'-{i:04}'+'.png', dpi=600)         
            plt.pause(0.00001)
            plt.clf()
            
        os.system('ffmpeg -framerate 60 -i '+path+basename+'-%04d.png -c:v libx264 -pix_fmt yuv420p '+path+basename+'.mp4')
        os.system('ffmpeg y-i '+path+basename+'.mp4 -vf "setpts=0.25*PTS" -r 60 '+path+basename+'-60.mp4')
        os.system('rm '+path+basename+'*.png')
        
    createImgs(path, basename, liquidityAll, feesAll, priceTicks, timeTicks, tsA)


    
```
